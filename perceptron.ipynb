# Perceptron Algorithm

```python
def display(weights, fig, ax, i):
    m = -(weights[1]/weights[2])
    b = -(weights[0]/weights[2])

    x_input = np.linspace(-3,3,100)
    y_input = m*x_input + b

    ax.plot(x_input, y_input, color='red', linewidth=1)
    ax.set_ylim(-3,2)
    ax.set_title(f"Iteration {i+1}")
    fig.show()
```

```python
def perceptron(X, y, plot_updates=False):
    if plot_updates:
        fig, ax = plt.subplots(figsize=(10,6))
        ax.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
        
    X_n = np.insert(X,0,1,axis=1)
    weights = np.ones(X_n.shape[1])
    lr = 0.01

    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X_n[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X_n[j]

        if plot_updates:
           display(weights, fig, ax, i)

    return weights[0],weights[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_, coef_ = perceptron(X,y,True)
print(f"Coeffcients : {coef_}")
print(f"Intercept : {intercept_}")
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```
